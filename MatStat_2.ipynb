2.1 Моделирование выборки случайных величин

```python
import numpy as np

def Geom(prob):
    j=0
    k = np.random.binomial(size=1, n=1, p=prob)
    while k==0:
        j=j+1
        k = np.random.binomial(size=1, n=1, p=prob)
    return j

def Neg_Binom(r, prob): #r, prob - параметры, задающие NB распределение
    xi = 0
    for i in range(r):
        xi += Geom(prob)
    return xi

def NewSample(sample, p, r):
    lstRes = np.array([])
    i = 1
    for i in range(sample):
        lstRes = np.append(lstRes, Neg_Binom(r, p))
    return lstRes
    
    
print(NewSample(5, 0.6, 100))# p = 0.6, r = 100
print(NewSample(10, 0.6, 100))
print(NewSample(100, 0.6, 100))
print(NewSample(1000, 0.6, 100))
print(NewSample(100000, 0.6, 100))
```

2.2 Построение эмпирической функции распределения

```python
import matplotlib.pyplot as plt
    
def Fact(n):
    global FactMas #массив, содержащий факториальные значения: в i-ой ячейке - i!.
    while((len(FactMas) - 1) < n):
        FactMas = np.append(FactMas, (FactMas[len(FactMas) - 1]) * len(FactMas))
    return FactMas[n]

FactMas = np.array([1, 1])

def Emp_Func(sample): # Общая функция, выводящая на экран графики эмпирической функции и функции распределения СВ
    x = np.array([])  # по заданной выборке и параметрам СВ
    y = np.array([0])
    x = np.arange(np.amax(sample) + 6)
    VarRad = np.sort(sample)
    k = 0
    for i in range (int(np.amax(sample)) + 1):
        if i < VarRad[k]:
            y = np.append(y, y[len(y) - 1])
        else:
            ch = 0
            while ((k <= len(VarRad) - 1) and (VarRad[k] == i)):
                k = k + 1
                ch = ch + 1
            y = np.append(y, y[len(y) - 1] + ch)
    for i in range(5):
        y = np.append(y, y[len(y) - 1])
    y = np.delete(y, y[0])
    y = y / y[len(y) - 1]
    #plt.ylim(0, 1.2)
    
    y2 = np.array([0])
    i = 0
    while(1 - y2[len(y2) - 1] > 0.018):
        c = int(3 + i - 1)
        y2 = np.append(y2, y2[len(y2) - 1] + (Fact(c) / (Fact(i) * Fact(3 - 1))) * 0.6**3 * (1-0.6)**i)  
        i = i + 1
    y1 = np.array([])
    for i in range(len(y2) - 1):#-2
        y1 = np.append(y1, y2[i + 1])
    if len(y1) < len(y):
        while(len(y1) < len(y)):
            y1 = np.append(y1, 1)
    else:
        while(len(y) < len(y1)):
            y = np.append(y, 1)
            x = np.append(x, x[len(x) - 1] + 1)
    
    plt.figure(figsize = (15, 8))
    plt.figsize=(30,20)
    plt.ylim(0, 1.2)
    plt.step(x, y, where = 'post', label = 'ЭФР')
    plt.step(x, y1, where = 'post', label = 'Функция распределения')
    plt.legend()
    plt.show()
```

```python
# Находим верхние границы разности для каждой пары эмпирических функций для каждого n:
    
def Emp_Func_Res(sample): 
    y = np.array([0])
    VarRad = np.sort(sample)
    k = 0
    for i in range (int(np.amax(sample)) + 1):
        if i < VarRad[k]:
            y = np.append(y, y[len(y) - 1])
        else:
            ch = 0
            while ((k <= len(VarRad) - 1) and (VarRad[k] == i)):
                k = k + 1
                ch = ch + 1
            y = np.append(y, y[len(y) - 1] + ch)
    for i in range(3):
        y = np.append(y, y[len(y) - 1])
    y = np.delete(y, y[0])
    y = y / y[len(y) - 1]
    return y
    
def Dif(mas1, mas2):
    i = 0
    maxim = 0
    if len(mas1) < len(mas2):
        l = len(mas1) - 1
    else:
        l = len(mas2) - 1
    for i in range(l):
        k = ((mas1[i] - mas2[i])**2)**0.5
        if maxim < k:
            maxim = k
    return maxim

print('Верхние границы разности Э.Ф.Р. при n = 5:')
print('Первая выборка:   ', Dif(Emp_Func_Res(e[0]), Emp_Func_Res(e[1])))
print('                  ', Dif(Emp_Func_Res(e[0]), Emp_Func_Res(e[2])))
print('                  ', Dif(Emp_Func_Res(e[0]), Emp_Func_Res(e[3])))
print('                  ', Dif(Emp_Func_Res(e[0]), Emp_Func_Res(e[4])))
```

2.4 Построение гистограммы и полигон частот.

```python
# Полигон частот:

def Poligon(sample): # sample - отсортированная по возрастанию выборка (вариационный ряд)
    k = 0
    ch = 0
    res = np.array([])
    for t in sample:
        if k == t:
            ch = ch + 1
        else:
            res = np.append(res, ch)
            ch = 0
            k = k + 1
            while k < t:
                k = k + 1
                res = np.append(res, 0)
            ch = ch + 1
    res = np.append(res, ch)
    return res # функция, возвращающая значения частот для заданного полигона
    
def PoligonDraw(sample):
    y = Poligon(np.sort(sample))
    x = np.arange(len(y))
    plt.figure(figsize = (15, 6))
    plt.plot(x, y)
    plt.scatter(x, y)
    plt.grid()
    plt.ylabel('Vr - \'частота\'')
    plt.xlabel('Ar - значения случайной величины')
    plt.show()
  
#Полигон частот для выборки с n = 5
PoligonDraw(e[0])
```

```python
# Строим гистограмму частот:

def Hist(sample, delta): # sample - отсортированная по возрастанию выборка (вариационный ряд), delta - интервал
    k = 1
    ch = 0
    res = np.array([])
    for t in sample:
        if k*delta >= t:
            ch = ch + 1
        else:
            res = np.append(res, ch)
            ch = 0
            k = k + 1
            while (k*delta) < t:
                k = k + 1
                res = np.append(res, 0)
            ch = ch + 1
    res = np.append(res, ch)
    return res # функция, возвращающая значения гистограммы при заданном интервале

def HistDraw(sample, delta):
    y = Hist(np.sort(sample), delta)
    y = y/(np.sum(y))
    x0 = np.arange(len(y))
    x = x0*delta
    plt.figure(figsize = (15, 6))
    plt.ylim = ((0, 1))
    plt.step(x, y, where = 'post')
    #plt.scatter(x, y)
    plt.grid()
    plt.ylabel('Vr* - относительная частота')
    plt.xlabel('Ar - значения случайной величины')
    plt.show()
    
# Гистограммы частот с разными интервалами для n = 5
HistDraw(e[0], 2)
HistDraw(e[0], 1)
HistDraw(e[0], 0.5) #Поскольку значения случайной величины принимают только целые значения, нет смысла использовать
                    #дробные интервалы для изучения гистограммы
```

```python
# Сравнение с функцией вероятности:

def ProbFunc(): # Функция, возвращающая значения функции вероятности для конечного числа неотрицательных значений xi
    y2 = np.array([1])
    i = 0
    while(y2[len(y2) - 1] > 0.0015):
        c = int(3 + i - 1)
        y2 = np.append(y2, (Fact(c) / (Fact(i) * Fact(3 - 1))) * 0.6**3 * (1-0.6)**i)  
        i = i + 1
    y1 = np.array([])
    for i in range(len(y2) - 1):
        y1 = np.append(y1, y2[i + 1]) #удаление нуля из начала
    return y1

def CompareDraw(sample1, sample2):
    sample1 = sample1/(np.sum(sample1))
    if len(sample2) > len(sample1):
        while len(sample2) > len(sample1):
            sample1 = np.append(sample1, 0)
    else:
        while len(sample1) > len(sample2):
            sample2 = np.append(sample2, 0)
    x = np.arange(len(sample1))
    plt.figure(figsize = (15, 6))
    plt.ylim = ((0, 1))
    plt.plot(x, sample1, color = 'red', linestyle = '-.', label = 'полигон относительных частот')
    plt.plot(x, sample2, color = 'blue', linestyle = '--', label = 'функция вероятности')
    plt.grid()
    plt.ylabel('Vr* - относительная частота / вероятность')
    plt.xlabel('Ar - значения случайной величины')
    plt.legend()
    plt.show()
    
#Проведем сравнение для n = 5
CompareDraw(Poligon(np.sort(e[4])), ProbFunc())
```

```python
# Сначала необходимо сгенерировать выборку (Y1,.., Yn), где Yi - случайная величина со стандартным нормальным распределением
import random

#print(NewSample(5, 0.6, 100))# p = 0.6, r = 100
#print(NewSample(10, 0.6, 100))
#print(NewSample(100, 0.6, 100))
#print(NewSample(1000, 0.6, 100))
#print(NewSample(100000, 0.6, 100))

def SampleNorm(sampleLen):# Функция генерирует выборку заданной длины из СВ с нормальным распределением
    mas = np.array([])
    a = 1
    b = 99999999999
    for i in range(sampleLen):
        mas = np.append(mas, (random.randrange(a, b, 1)/100000000000))
    return mas

def SampleRel(sample, sig):# Функция генерирует выборку СВ с распределением Релея по выборке с СВ, распределенными нормально
    res = np.array([])
    for t in sample:
        res = np.append(res, math.sqrt(math.log(1 - t)*(-1)*2*(sig**2)))
    return res

#eNorm = np.array([])# Строим выборку для n = 5:
#№eNorm = SampleNorm(5)
#№eRel = SampleRel(eNorm, 100)#Параметр масштаба возьмем равным 100
#print(eRel)

print(SampleRel(SampleNorm(5), 100))
print(SampleRel(SampleNorm(10), 100))
print(SampleRel(SampleNorm(100), 100))
print(SampleRel(SampleNorm(1000), 100))
print(SampleRel(SampleNorm(100000), 100))
#x = np.array([])
#for i in range(100):
#    x = np.append(x,i)
#plt.plot(x, aRel)
#plt.show()
```

```python
# Полигон частот:

def PoligonRel(sample): # sample - отсортированная по возрастанию выборка (вариационный ряд)
    k = 0
    ch = 0
    res = np.array([])
    for i in range(len(sample) - 1):
        sample[i] = int(sample[i])
    for t in sample:
        if k == t:
            ch = ch + 1
        else:
            res = np.append(res, ch)
            ch = 0
            k = k + 1
            while k < t:
                k = k + 1
                res = np.append(res, 0)
            ch = ch + 1
    res = np.append(res, ch)
    return res # функция, возвращающая значения частот для заданного полигона
    
def PoligonDrawRel(sample):
    y = PoligonRel(np.sort(sample))
    x = np.arange(len(y))
    plt.figure(figsize = (15, 6))
    plt.plot(x, y)
    plt.scatter(x, y)
    plt.grid()
    plt.ylabel('Vr - \'частота\'')
    plt.xlabel('Ar - значения случайной величины')
    plt.show()
  
#Полигон частот для выборки с n = 5
PoligonDrawRel(eRel[0])
```
